```python
NEO4J_URI = "bolt://<uri>:7687"
```

```python
from graphdatascience import GraphDataScience

# Use Neo4j URI and credentials according to your setup
# NEO4J_URI could look similar to "bolt://my-server.neo4j.io:7687"
gds = GraphDataScience(NEO4J_URI, auth=("Daniela", "password"), aura_ds=False)

# Check the installed GDS version on the server
print(gds.version())
assert gds.version()
```

```python
gds.run_cypher("""
MATCH (a:Account)
WITH a LIMIT 1
OPTIONAL MATCH (a)-[r:FROM|TO]-(t:Transaction)
RETURN a{.a_id,.name,.email} AS account,
    CASE type(r) WHEN "FROM" THEN "sent" ELSE "was sent" END AS verb,
    t.amount AS amount;
""")
```
